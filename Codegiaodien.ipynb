```python
from tkinter import *
from PIL import Image, ImageTk
from tkinter import ttk
from tkinter.filedialog import *
from tkinter.messagebox import *
import time
import numpy as np
import os
from tensorflow.keras.utils import load_img, img_to_array
from keras.models import load_model
import matplotlib.pyplot as plt
file =''
Label1 = ''
model_CNN = ''
os.environ['TF CPP MIN LOG LEVEL'] = '2'
model_CNN = load_model('Nhandiensaubenh.h5')
wd1 = Tk()
wd1.geometry('700x500+370+80')
wd1.title('Plant disease')
wd1.iconbitmap('hinhanh/sauxanh.ico')
wd1.resizable(False,False)
canvas_wd1 = Canvas(wd1, width=700, height=500)
bg_wd1 = Image.open('hinhanh/saubenhhaicay.jpg')
n_bg_wd1 = bg_wd1.resize((700,500),Image.Resampling.LANCZOS)
open_n_bg_wd1 = ImageTk.PhotoImage(n_bg_wd1)
canvas_wd1.create_image(0, 0, anchor=NW, image=open_n_bg_wd1)
canvas_wd1.pack()
def open_wd2():
	wd2 = Toplevel(wd1)
	wd2.title('Plant Disease')
	wd2.iconbitmap('hinhanh/sauxanh.ico')
	wd2.geometry('1000x500+330+0')
	wd2.resizable(False, False)
	LB_wd2 = LabelFrame(wd2,bg = 'sky blue', width=1000, height=1000)
	LB_wd2.pack()
	fr_wd2 = Frame(LB_wd2,bg = 'dark turquoise', width = 500,height = 400)
	fr_wd2.place(x=400,y=50)
	Label2 = Label(LB_wd2,text = "Result:",font = ('Arial',16),bg = 'DarkOrange1').place( x = 10, y = 10)
	Label2_value = Entry(LB_wd2,width = 20,font = ('Arial',16))
	Label2_value.place( x = 100, y = 10)
	wd1.withdraw()
	wd2.deiconify()
	def open_file():
		cancel()
		global file,Label1
		file = askopenfilename(defaultextension = '.jpg',filetypes = [('All Files','*.*'),('JPG Files','*.jpg'),('PNG Files','*.png')])
		if file == '':
			showinfo('Notice','Please choose a photo!')
		else:
			openfile = Image.open(file)
			n_openfile = openfile.resize((485,400), Image.Resampling.LANCZOS)
			showfile = ImageTk.PhotoImage(n_openfile)
			Label1 = Label(fr_wd2,image = showfile)
			Label1.pack()
			mainloop()
	def cancel():
		global Label1,Frame1,file
		if Label1 == '':
			return
		else:
			Label1.destroy()
		file = ''
		fr_wd2.configure()
		Label2_value.delete(0,END)
	def check():
		global file,model_CNN
		if file =='':
			showinfo('Notice','Please choose a photo!')
		else:
			img = load_img(file,target_size =(150,150))
			img = img_to_array(img)
			img = img.astype('float32')
			img = img/255
			img = np.expand_dims(img,axis=0)
			result = model_CNN.predict(img)
			classname = ['Ray phan', 'Sau an bong',"Sau duc than",'Sau duc trai']
			for i in range (0,4):
				if round(result[0][i])== 1:
					prediction = classname[i]
			a = f"This is '{prediction}'"
			Label2_value.insert(0,a)
	button_check=Button(LB_wd2,bg = 'LightSteelBlue4',text='CHECK',command = check, width=20, height=2)
	button_check.place(x=100,y=200)
	button_choose=Button(LB_wd2,bg = 'MediumOrchid4',text='CHOOSE',command = open_file, width=20, height=2)
	button_choose.place(x=100,y=100)
	button_cancel = Button(LB_wd2,bg = 'IndianRed3', text='CANCEL', command=cancel, width=20, height=2)
	button_cancel.place(x=100, y=300)
button_next = Button(wd1,bg = 'cornsilk2', text='GET GO', command=open_wd2, width=25, height=2)
button_next.place(x=250, y=450)

mainloop()
```
